```python
import numpy as np
import matplotlib.pyplot as plt
```

1-3

```python
for N in [1, 2, 10, 20]:
    x_means = []

    for _ in range(10000):
        xs = []
        for n in range(N):
            x = np.random.rand()
            xs.append(x)
        
        mean = np.mean(xs)
        x_means.append(mean)

    plt.hist(x_means, label=f"{N}", bins=100, alpha=0.7)

plt.legend(loc="upper right")
plt.show()
```

1-4

```python
for N in [1, 2, 10, 20]:
    x_sums = []

    for _ in range(10000):
        xs = []
        for n in range(N):
            x = np.random.rand()
            xs.append(x)
        
        sum = np.sum(xs)
        x_sums.append(sum)

    plt.hist(x_sums, label=f"{N}", bins=100, alpha=0.7)
    plt.plot()

plt.legend(loc="upper right")
plt.show()
```
