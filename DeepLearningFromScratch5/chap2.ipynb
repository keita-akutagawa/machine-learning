```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

2-2

```python
xs = np.loadtxt("height.txt")

plt.hist(xs, bins=100, density=True)
plt.show()
```

```python
mu = np.mean(xs)
sigma = np.std(xs)

print(mu, sigma)
```

```python
x = np.linspace(150, 190, 1000)
y = 1.0 / np.sqrt(2.0 * np.pi) / sigma * np.exp(-(x - mu)**2 / (2.0 * sigma**2))

plt.hist(xs, bins=100, density=True)
plt.plot(x, y)
plt.show()
```

2-4

```python
samples = np.random.normal(mu, sigma, size=10000)

plt.hist(xs, bins=100, density=True, alpha=0.7, label="original data")
plt.hist(samples, bins=100, density=True, alpha=0.7, label="generated data")
plt.legend(loc='upper right')
plt.show()
```

```python
p1 = stats.norm.cdf(160, mu, sigma)
p2 = stats.norm.cdf(200, mu, sigma)

print("under 160cm : ", p1)
print("upper 200cm : ", 1.0 - p2)
```
